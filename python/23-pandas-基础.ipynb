### series

```python
data = {'country': ['aaa', 'bbb', 'ccc'], 'population': [10, 11, 12]}
df_data = pd.DataFrame(data)
df_data
```

```python
p = df_data['population']
print(p)
print(type(p))
print(p.mean())
print(p.max())
print(p.min())
```
